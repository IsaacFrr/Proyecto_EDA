##### Columnas clave del proyecto

```python
# Vamos a verificar que las columnas que necesitamos para las 3 dimensiones existen en ambos años
cols_clave = ['RemoteWork', 'ConvertedCompYearly', 'YearsCodePro',
              'LanguageHaveWorkedWith', 'Country']

print("=== 2022 ===")
for col in cols_clave:
    existe = col in df_2022.columns
    print(f"  {col}: {'✓ existe' if existe else '✗ NO existe'}")

print("\n=== 2025 ===")
for col in cols_clave:
    existe = col in df_2025.columns
    print(f"  {col}: {'✓ existe' if existe else '✗ NO existe'}")
```

```python
# Buscar columnas que puedan equivaler a 'YearsCodePro' en 2025 (Mide los años de experiencia)
print("\n=== Columnas con 'code' en 2025 ===")
print([c for c in df_2025.columns if 'code' in c.lower()])

print("\n=== Columnas con 'work' en 2025 ===")
print([c for c in df_2025.columns if 'work' in c.lower()])

print("\n=== Columnas con 'exp' en 2025 ===")
print([c for c in df_2025.columns if 'exp' in c.lower()])
```

```python
# Antes de renombrar, hay que asegurarse de que ambas columnas miden lo mismo.

# Ver qué tipo de valores tiene WorkExp en 2025
print("=== WorkExp (2025) ===")
print(df_2025['WorkExp'].describe())
print(f"\nTipo: {df_2025['WorkExp'].dtype}")
print(f"\nMuestra: {df_2025['WorkExp'].head(10).tolist()}")

# Comparar con YearsCodePro de 2022
print("\n=== YearsCodePro (2022) ===")
print(df_2022['YearsCodePro'].describe())
print(f"\nTipo: {df_2022['YearsCodePro'].dtype}")
print(f"\nMuestra: {df_2022['YearsCodePro'].head(10).tolist()}")
```
